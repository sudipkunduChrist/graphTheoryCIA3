```python
import heapq
import time
from rich.console import Console
from rich.table import Table
from rich.panel import Panel
from rich.live import Live
from rich.progress import SpinnerColumn, TextColumn

console = Console()

class CampusGraphEngine:
    def __init__(self):
        # UNIT 3 CONCEPT: Adjacency List Representation
        # Nodes = Locations, Values = (Neighbor, Distance in Meters)
        self.graph = {
            'Main Gate': [('Block A', 100), ('Block B', 150), ('Dominos', 50)],
            'Block A': [('Main Gate', 100), ('Cafeteria', 30), ('Freshiteria', 60)],
            'Block B': [('Main Gate', 150), ('B-Basement', 20), ('Rooftop', 80), ('Block A', 50)],
            'Cafeteria': [('Taste of Dilli', 5), ('Punjabi Bites', 5), ('Rolls Lane', 5),
                          ('Southern Delights', 5), ('Bites and Brews', 5)],
            'B-Basement': [('Nescafe', 5)],
            'Rooftop': [('Gianis', 5), ('Amritsari Haveli', 5)],
            'Freshiteria': [('Greenox', 5)],
            'Dominos': [('Main Gate', 50)]
        }
        self.outlets = ['Taste of Dilli', 'Punjabi Bites', 'Rolls Lane', 'Southern Delights',
                        'Bites and Brews', 'Nescafe', 'Gianis', 'Amritsari Haveli', 'Greenox', 'Dominos']

    def find_shortest_path(self, start_node):
        # UNIT 3 CONCEPT: Dijkstra's Algorithm (Single Source Shortest Path)
        distances = {node: float('inf') for node in self.graph}
        distances[start_node] = 0
        pq = [(0, start_node)]

        while pq:
            current_dist, u = heapq.heappop(pq)
            if current_dist > distances.get(u, float('inf')):
                continue
            for v, weight in self.graph.get(u, []):
                if current_dist + weight < distances[v]:
                    distances[v] = current_dist + weight
                    heapq.heappush(pq, (distances[v], v))
        return distances
```
